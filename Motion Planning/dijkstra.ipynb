# Dijkstra's Algorithm

This is a greedy algorithm which finds the shortest path in a directed/non-directed graph. Lets start by implementing it. First create a graph structure to store the graph vertices and edge values. And then write the algorithm to access the graph.

```python
import numpy as np

vertices = np.array([1, 2, 3, 4, 5, 6])
edges = [[2, 3], [3, 4], [5], [6], [4, 6], []]
costs = [[2, 4], [1, 7], [3], [1], [2, 5], []]

svertex = vertices[0] # starting vertex
gvertex = vertices[-1] # goal vertex
selected = []
```

```python
cvertex = svertex
while len(selected) < 5:

    # start with the current vertex: and access all the connections:
    tables = []
    table = []
    for each in vertices:
        if each == cvertex or each in selected:
            continue
        else:
            try:
                table.append(costs[np.where(vertices == cvertex)[0][0]][np.where(vertices == each)[0][0]])
            except:
                table.append(np.inf)
    selected.append(0)
    
    print(table)

```
